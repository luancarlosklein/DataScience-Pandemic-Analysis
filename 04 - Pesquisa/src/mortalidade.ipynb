## Realizando o Ajuste de mortalidade de acordo com as idades

Primeiro importando as bibliotecas necessárias

```python
import pandas as pd
```

Após importamos o dataset que contém dos dados de mortes, proveninentes do SivepGripe

```python
df = pd.read_csv(r'mortes_idade.csv',sep=',')
df
```

Após realizamos a importação dos dados demográficos, provenientes do IBGE

```python
df_ibge = pd.read_excel('IDH_2010.xls', sep=';')
df_ibge
```

O próximo passo consiste em selecionar as colunas nas quais estamos interessados do IBGE, e renomeamos elas.
Primeiro definimos para a população masculina.

```python
## Selecionando as colunas das populacoes masculinas por faixa etaria
df_idades = df_ibge[['Município', 'Código do Município', 'População masculina de 0 a 4 anos', 'População masculina de 10 a 14 anos', 'População masculina de 15 a 19 anos',
                   'População masculina de 20 a 24 anos', 'População masculina de 25 a 29 anos', 'População masculina de 30 a 34 anos',
                   'População masculina de 35 a 39 anos', 'População masculina de 40 a 44 anos', 'População masculina de 45 a 49 anos',
                   'População masculina de 50 a 54 anos', 'População masculina de 55 a 59 anos', 'População masculina de 5 a 9 anos',
                   'População masculina de 60 a 64 anos', 'População masculina de 65 a 69 anos', 'População masculina de 70 a 74 anos',
                   'População masculina de 75 a 79 anos', 'População masculina com 80 anos e mais', 'População total']]

## Processo de renomear
tmp = 0
for i in range(len(df_idades.columns)):
    if (not(i == 0 or i == 1 or tmp > 80)):
        df_idades.rename(columns={ df_idades.columns[i]: tmp }, inplace = True)
        tmp += 5

df_idades
```

Fazemos o mesmo processo, porém para a população feminina.

```python
## Selecionando as colunas das populacoes feminina por faixa etaria
df_idades_f = df_ibge[['Município', 'Código do Município', 'População feminina de 0 a 4 anos', 'População feminina de 10 a 14 anos', 'População feminina de 15 a 19 anos',
                   'População feminina de 20 a 24 anos', 'População feminina de 25 a 29 anos', 'População feminina de 30 a 34 anos',
                   'População feminina de 35 a 39 anos', 'População feminina de 40 a 44 anos', 'População feminina de 45 a 49 anos',
                   'População feminina de 50 a 54 anos', 'População feminina de 55 a 59 anos', 'População feminina de 5 a 9 anos',
                   'População feminina de 60 a 64 anos', 'População feminina de 65 a 69 anos', 'População feminina de 70 a 74 anos',
                   'População feminina de 75 a 79 anos', 'População feminina com 80 anos e mais', 'População total']]

## Processo de renomear
tmp = 0
for i in range(len(df_idades_f.columns)):
    if (not(i == 0 or i == 1 or tmp > 80)):
        df_idades_f.rename(columns={ df_idades_f.columns[i]: tmp }, inplace = True)
        tmp += 5

df_idades_f
```

Agora somamos os dois dataframes das duas populações, masculina e feminina.

```python
# Somando as populacoes masculinas e femininas por faixa etaria
df_pop = df_idades.copy()
tmp = 0
for i in range (len(df_idades.columns)):
    if (not(i==0 or i == 1 or tmp > 80)):
        total = df_idades[tmp] + df_idades_f[tmp]
        df_pop[tmp] = total
        tmp += 5 
df_pop
```

Criamos uma nova coluna no dataframe inicial para armazenar as populações

```python
df['POPULACAO'] = 0
df
```

Renomeia o nome das colunas do primeiro dataframe, para casar com o dataframe gerado anteriormente

```python
# Renomeando a coluna das faixas etarias
tmp = 0
for index, row in df.iterrows():
    df.at[index, 'Idade'] = tmp
    tmp += 5
    if (tmp > 80):
        tmp = 0
df
```

Adiciona a população em cada faixa etaria

```python
for index, row in df.iterrows():
    cod = row['CO_MUN_NOT']
    pop = df_pop.loc[df_pop['Código do Município'] == cod, row['Idade']]
    if (len(pop) > 0):
        df.at[index, 'POPULACAO'] = pop.iloc[0]
df
```

Cria o dataframe que irá conter os dados finais

```python
df_mort = df_pop.copy()
df_mort = df_mort[['Município', 'Código do Município', 'População total']]
df_mort['Mortalidade'] = 0
df_mort
```

Realiza o calculo da mortalidade ajustada pela idade, de acordo com o definido pela WHO. 

```python
fx = 0

## Constantes definidas pelo artigo do WHO
cte = [8.86, 8.69, 8.60, 8.47, 8.22, 7.93, 7.61, 7.15, 6.59, 6.04, 5.37, 4.55, 3.72, 2.96, 2.21, 1.52, 1.54]
for index, row in df.iterrows():
    cod = row['CO_MUN_NOT']
    tmp = df_pop.loc[df_pop['Código do Município'] == cod, 'População total']
    if (len(tmp) > 0):
        pop_tot = tmp.iloc[0]
    else:
        pop_tot = 0
    mort = df.at[index, 'Casos'] * (cte[fx] / df.at[index, 'POPULACAO'])
    cid_m = df_mort.loc[df_mort['Código do Município'] == cod, 'Mortalidade']
    i = df_mort.index[df_mort['Código do Município'] == cod].tolist()
    mort += cid_m
    df_mort.at[i, 'Mortalidade'] = mort
    fx += 1
    if (fx > 16):
        fx = 0
df_mort
```

Cria uma coluna para a população idosa

```python
df_mort.loc[df_mort['Mortalidade'] == 0]
df_mort = df_mort.sort_values('Mortalidade')
df_mort['POPULACAO IDOSA'] = 0
df_mort
```

Consideramos aqui a população idosa aquela acima de 60 anos.
Então realizamos o cálculo para somar todos os idosos de uma cidade.

```python
## Realiza a soma da população acima de 60 anos
for index, row in df_mort.iterrows():

    cod = row['Código do Município']
    pop = df_pop.loc[df_pop['Código do Município'] == cod]
    tmp = df_pop.loc[df_pop['Código do Município'] == cod, 'População total']
    
    if (len(tmp) > 0):
        pop_tot = tmp.iloc[0]
    else:
        pop_tot = 0
    
    if (len(pop) > 0):
        aux = pop.iloc[0][14:19]
        
        for i in aux:
            
            tmp2 = df_mort.at[index, 'POPULACAO IDOSA'] + i
            df_mort.at[index, 'POPULACAO IDOSA'] = tmp2
            
df_mort
```

Por fim cria uma coluna com o percentual da população de cada cidade,
e salva em um dataset.

```python
df_mort['POPULACAO IDOSA PERCENTUAL'] = df_mort['POPULACAO IDOSA'] / df_mort['População total']
df_mort.to_csv('mortalidadeAjustada.csv')
```
